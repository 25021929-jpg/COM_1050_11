```python
import math

## Hàm 1: Kiểm tra tính nguyên tố (is_prime)
def is_prime(n):
    """
    Kiểm tra xem một số nguyên dương n có phải là số nguyên tố hay không.
    Sử dụng phương pháp chia thử tối ưu (chỉ kiểm tra đến căn bậc hai của n).
    """
    if n < 2:
        return False
    
    # Chỉ cần kiểm tra đến căn bậc hai của n
    # Thêm 1 để đảm bảo vòng lặp bao gồm số nguyên lớn nhất <= sqrt(n)
    limit = int(math.sqrt(n)) + 1 
    
    for i in range(2, limit):
        if n % i == 0:
            return False
            
    return True

## Hàm 2: Tính tổng các số nguyên tố trong đoạn [a, b] (sum_primes)
def sum_primes(a, b):
    """
    Tính tổng tất cả các số nguyên tố trong đoạn [a, b].
    """
    if a > b:
        # Đảm bảo a <= b
        a, b = b, a 
    
    tong_so_nguyen_to = 0
    
    # Duyệt qua tất cả các số trong đoạn [a, b]
    for num in range(a, b + 1):
        if is_prime(num):
            tong_so_nguyen_to += num
            
    return tong_so_nguyen_to
a = 10
b = 50

print(sum_primes(a, b))
```

```python
def is_palindrome(s):
  s = str(s)
  return s == s[::-1] #Kiểm tra s có bằng số đảo ngược của nó không
def reversed(n):
  ans = 0
  while n !=0:
    ans = ans *10 + n%10
    n//=10
  return ans
def reverse_add_palindrome(s):
  count = 0
  while is_palindrome(s) == False:
    s += reversed(s)
    count += 1
  print(s,count,sep = ' ')
s = int(input())
reverse_add_palindrome(s)
```

```python
#Nhập vào chuỗi a và chuỗi b. Hãy xóa chuỗi b trong chuỗi a rồi in lại chuỗi a trong lên (không được dùng hàm replace)
A = input()
B = input()
def xoa_chuoi(A,B):
  vi_tri_thoa_man = 0
  check = False
  #Sử dụng cách thủ công (Khi nào đến một chữ giống chữ của b thì bắt đầu xét substring của a có độ dài là b. Nếu giống thì ok)
  #Duyệt từ 0 cho đến index cuối cùng mà A có thể tạo thành B
  for i in range(len(A) - len(B) + 1):
    if A[i] != B[0]:
      continue
    temp = A[i:i+len(B)] #A[i] là chữ cái đầu tiên của substring đang test
    if temp == B:
      vi_tri_thoa_man = i
      check = True
      break
  if check:
    A = A[:vi_tri_thoa_man] + A[vi_tri_thoa_man + len(B):]
  print(A)
xoa_chuoi(A,B)
```

```python
def fibon(n):
  #Sử dụng cộng dồn dp
  prev = 1 #F(i-1)
  prevprev = 0 #F(i-2)
  current = prev + prevprev 
  while current <= n:
    prevprev = prev
    prev = current
    current = prev + prevprev
    print(current,end = ' ')
  return prev #Vì ở lúc cuối thì current đã > n
n = int(input())
print(fibon(n))
```

```python
#Nhập vào n in ra các chữ số chính phương <=n có các chữ số đôi một khác nhau (in ra theo thứ tự tăng đần,cùng hàng,cách nhau bởi một dấu cách). Nếu không có in ra no number
from math import isqrt
def kiem_tra_doi_mot_khac_nhau(n):
  #Sử dụng set là nhanh nhất
  n = str(n) #Chuyển thành chuỗi để dễ dàng duyệt qua từng chữ số
  # Sử dụng set để kiểm tra tính duy nhất
    # len(set(s_num)) sẽ trả về số lượng chữ số duy nhất
    # len(s_num) sẽ trả về tổng số lượng chữ số
  return len(set(n)) == len(n) #Set sẽ không cho phép việc có 2 giá trị giống nhau
def so_chinh_phuong(n):
  ans = []
  for i in range(1,isqrt(n)+1):
    so_chinh_phuong = i*i
    if kiem_tra_doi_mot_khac_nhau(so_chinh_phuong):
      ans.append(str(so_chinh_phuong))
  if ans:
    print(' '.join(ans)) #Chỉ có thể dùng hàm join với list có các element là str 
  else:
    print('no number')
n = int(input())
so_chinh_phuong(n)
```
